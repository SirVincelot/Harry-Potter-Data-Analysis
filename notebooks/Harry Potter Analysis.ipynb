```python
# import packages
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
df = pd.read_csv("Characters.csv", sep=";")
df.head(3)
```

```python
gender_equality = df.groupby("Gender").Name.count().reset_index()
gender_equality.columns = ["Gender", "Count"]
gender_equality.head()
```

```python
hp_colors = color_list = ['#740001', '#ae0001', '#eeba30', '#d3a625', '#5d5d5d', '#aaaaaa', '#0e1a40', '#bebebe', '#946b2d', '#ecb939', '#f0c75e', '#726255', '#372e29']

plt.pie(gender_equality["Count"], labels = gender_equality["Gender"], colors = hp_colors);
plt.title("Gender distribution in Harry Potter");
```

```python
houses = df.groupby("House").Name.count().reset_index()
houses.columns = ["House", "Count"]
houses
```

```python
plt.pie(houses["Count"], labels = houses["House"], colors = hp_colors);
plt.title("House distribution in Harry Potter");
```
